## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

The first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

## Data Preprocessing

```python
# Client data preprocessing

# Convert date features to correct types
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'], format='%Y-%m-%d')
client_df['date_end'] = pd.to_datetime(client_df['date_end'], format='%Y-%m-%d')
client_df['date_modif_prod'] = pd.to_datetime(client_df['date_modif_prod'], format='%Y-%m-%d')
client_df['date_renewal'] = pd.to_datetime(client_df['date_renewal'], format='%Y-%m-%d')

# Convert categorical features to correct types
client_df['has_gas'] = client_df['has_gas'].astype('category')
client_df['origin_up'] = client_df['origin_up'].astype('category')
client_df['channel_sales'] = client_df['channel_sales'].astype('category')

# Check info again
client_df.info()
```

```python
# Price data preprocessing

# Convert date features to correct types
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')

price_df.info()
```

## EDA

```python
# Plot the churn status of the company
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

From this plot we know the dataset is imbalanced between churned and retained classes. Resampling techniques might needed to make the dataset more balanced.

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
axs.set_title('Distribution of Electricity Consumption in 12 Months')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_gas_12m', axs)
axs.set_title('Distribution of Gas Consumption in 12 Months')
plt.show()
```

```python
# Check out the categorical features that are strongly associated with churn
categorical = client_df[['has_gas', 'churn']]
fig, ax = plt.subplots(nrows=1, figsize=(18, 5))
sns.countplot(data=categorical, x='has_gas', hue='churn', ax=ax)
ax.set(title='Whether the client has gas or not')
plt.show()
```

```python
# Cehck for temporal features

# Create client tenure in months (or days/years) from a reference point
client_df['tenure_months'] = (client_df['date_end'].dt.to_period('M') - client_df['date_activ'].dt.to_period('M')).apply(lambda x: x.n)

temporal = client_df[['date_activ', 'date_end', 'date_modif_prod', 'date_renewal', 'tenure_months', 'churn']]

client_df['tenure_bin'] = pd.cut(client_df['tenure_months'], bins=[24, 36, 60, 120, 180])
churn_rate_by_bin = client_df.groupby('tenure_bin')['churn'].mean().reset_index()

sns.barplot(x='tenure_bin', y='churn', data=churn_rate_by_bin)
```

As expected, the longer the contract tenure is, the less possibility for a client to churn. This tells us the temporal effect matters, and we probably need to segment the users based on their tenure in order to better capture different behaviors.
